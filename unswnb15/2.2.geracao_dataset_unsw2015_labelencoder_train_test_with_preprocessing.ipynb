# Pré-processamento com Condificação (LabelEncoder)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
pd.set_option('display.max_rows', None)
print(df_unsw_full.dtypes)
pd.set_option('display.max_rows', False)
```

```python
df_unsw_full.head()
```

```python
# cria variáveis recebendo a função LabelEncoder
label_encoder_sport  = LabelEncoder()
label_encoder_dsport = LabelEncoder()
label_encoder_proto  = LabelEncoder()
label_encoder_state = LabelEncoder()
label_encoder_service = LabelEncoder()
```

```python
# faz a codificação das variáveis nominais (colunas) específicas
#srcip #sport #dstip #dsport #proto #state #service
df_unsw_full['sport'] = label_encoder_proto.fit_transform(df_unsw_full['sport'])
df_unsw_full['dsport'] = label_encoder_proto.fit_transform(df_unsw_full['dsport'])
df_unsw_full['proto'] = label_encoder_proto.fit_transform(df_unsw_full['proto'])
df_unsw_full['state'] = label_encoder_state.fit_transform(df_unsw_full['state'])
df_unsw_full['service'] = label_encoder_service.fit_transform(df_unsw_full['service'])
```

# Divisão dataset full em Train (50%) e Test - Base-Line (50%)

```python
# separação dos tipos de ataque (coluna attack_cat)
df_benign = df_unsw_full[df_unsw_full.attack_cat == 'Benign']
df_generic = df_unsw_full[df_unsw_full.attack_cat == 'Generic']
df_exploits = df_unsw_full[df_unsw_full.attack_cat == 'Exploits']
df_fuzzers = df_unsw_full[df_unsw_full.attack_cat == 'Fuzzers']
df_dos = df_unsw_full[df_unsw_full.attack_cat == 'DoS']
df_reconnaissance = df_unsw_full[df_unsw_full.attack_cat == 'Reconnaissance']
df_analysis = df_unsw_full[df_unsw_full.attack_cat == 'Analysis']
df_backdoor = df_unsw_full[df_unsw_full.attack_cat == 'Backdoor']
df_shellcode = df_unsw_full[df_unsw_full.attack_cat == 'Shellcode']
df_worms = df_unsw_full[df_unsw_full.attack_cat == 'Worms']
```
