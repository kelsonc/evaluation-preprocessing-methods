# Análise e tratamento das colunas

```python
# renomeando os nomes dos Label
df_cic.loc[df_cic.Label == 'FTP-Patator', df_cic.Label.name] = 'BruteForce'
df_cic.loc[df_cic.Label == 'SSH-Patator', df_cic.Label.name] = 'BruteForce'

df_cic.loc[df_cic.Label == 'DoS slowloris', df_cic.Label.name] = 'DoS'
df_cic.loc[df_cic.Label == 'DoS Slowhttptest', df_cic.Label.name] = 'DoS'
df_cic.loc[df_cic.Label == 'DoS Hulk', df_cic.Label.name] = 'DoS'
df_cic.loc[df_cic.Label == 'DoS GoldenEye', df_cic.Label.name] = 'DoS'

df_cic.loc[df_cic.Label == 'Web Attack � Brute Force', df_cic.Label.name] = 'WebAttack'
df_cic.loc[df_cic.Label == 'Web Attack � XSS', df_cic.Label.name] = 'WebAttack'
df_cic.loc[df_cic.Label == 'Web Attack � Sql Injection', df_cic.Label.name] = 'WebAttack'

df_cic.loc[df_cic.Label == 'Bot', df_cic.Label.name] = 'BotNet'
```
