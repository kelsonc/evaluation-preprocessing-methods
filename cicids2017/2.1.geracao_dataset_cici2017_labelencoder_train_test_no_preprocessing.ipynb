# Divisão dataset full em Train (50%) e Test - Base-Line (50%)

```python
# separação dos tipos de ataque (coluna Label)
df_BENIGN = df_cic_full[df_cic_full.Label_multiclass == 'BENIGN']
df_DoS = df_cic_full[df_cic_full.Label_multiclass == 'DoS']
df_PortScan = df_cic_full[df_cic_full.Label_multiclass == 'PortScan']
df_DDoS = df_cic_full[df_cic_full.Label_multiclass == 'DDoS']
df_BruteForce = df_cic_full[df_cic_full.Label_multiclass == 'BruteForce']
df_WebAttack = df_cic_full[df_cic_full.Label_multiclass == 'WebAttack']
df_BotNet = df_cic_full[df_cic_full.Label_multiclass == 'BotNet']
df_Infiltration = df_cic_full[df_cic_full.Label_multiclass == 'Infiltration']
df_Heartbleed = df_cic_full[df_cic_full.Label_multiclass == 'Heartbleed']
```

```python
# dimensão do dataframe fragmentado
print(df_BENIGN.shape)
print(df_DoS.shape)
print(df_PortScan.shape)
print(df_DDoS.shape)
print(df_BruteForce.shape)
print(df_WebAttack.shape)
print(df_BotNet.shape)
print(df_Infiltration.shape)
print(df_Heartbleed.shape)
```

```python
# reseta os índices
df_BENIGN.reset_index(drop=True, inplace=True)
df_DoS.reset_index(drop=True, inplace=True)
df_PortScan.reset_index(drop=True, inplace=True)
df_DDoS.reset_index(drop=True, inplace=True)
df_BruteForce.reset_index(drop=True, inplace=True)
df_WebAttack.reset_index(drop=True, inplace=True)
df_BotNet.reset_index(drop=True, inplace=True)
df_Infiltration.reset_index(drop=True, inplace=True)
df_Heartbleed.reset_index(drop=True, inplace=True)
```

```python
# divisão dos dataframe fragmentados por linha (50 a 50)
df1_BENIGN50 = df_BENIGN.iloc[:1136548,:]
df2_BENIGN50 = df_BENIGN.iloc[1136549:,:]
```

```python
# juntando os dataframes train50
df_cic_train50 = pd.concat([df1_BENIGN50, df1_DoS50, df1_PortScan50, df1_DDoS50, df1_BruteForce50, 
                            df1_WebAttack50, df1_BotNet50, df1_Infiltration50, df1_Heartbleed50])
# embaralha do dataframe 
df_cic_train50 = shuffle(df_cic_train50)
# reseta os indices
df_cic_train50.reset_index(drop=True, inplace=True)
# verifica a dimensão
df_cic_train50.shape
```
